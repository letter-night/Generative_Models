```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

# Goal: Sample from bivariate Normal

```python
automatic_samples = np.random.multivariate_normal([0,0],[[1,0.5],[0.5,1]], 10000)

plt.scatter(automatic_samples[:,0], automatic_samples[:,1], s=0.9);
```

# Gibbs Sampling

```python
samples = {'x':[1], 'y':[-1]}

num_samples = 10000

for _ in range(num_samples):
    curr_y = samples['y'][-1]
    new_x = np.random.normal(curr_y/2, np.sqrt(3/4))
    new_y = np.random.normal(new_x/2, np.sqrt(3/4))
    samples['x'].append(new_x)
    samples['y'].append(new_y)
```

```python
plt.scatter(samples['x'], samples['y'], s=0.9);
```

```python
plt.hist(automatic_samples[:,0], bins=20, density=True, alpha=0.4)
plt.hist(samples['x'], bins=20, density=True, alpha=0.4);
```

```python
plt.hist(automatic_samples[:,1], bins=20, density=True, alpha=0.4)
plt.hist(samples['y'], bins=20, density=True, alpha=0.4);
```

```python
plt.scatter(automatic_samples[:-1, 0], automatic_samples[1:, 0], s=0.9)
print(pearsonr(automatic_samples[:-1, 0], automatic_samples[1:, 0])[0])
```

```python
plt.scatter(samples['x'][:-1], samples['x'][1:], s=.9)
print(pearsonr(samples['x'][:-1], samples['x'][1:])[0])
```
